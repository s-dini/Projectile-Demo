```python
import matplotlib.pyplot as plt
import numpy as np

def plot_graph(x, y, title="Graph", xlabel="X-axis", ylabel="Y-axis", line_style="-", color="blue", marker=None):
    """
    Plots a graph using the provided x and y values.

    Parameters:
    - x: list of x values
    - y: list of y values
    - title: title of the graph (default: "Graph")
    - xlabel: label for the x-axis (default: "X-axis")
    - ylabel: label for the y-axis (default: "Y-axis")
    - line_style: style of the plot line (default: "-") could be solid or dashed or dotted
    - color: color of the line (default: "blue")
    - marker: marker style (default: None), defines how individual data points look like (shape) 
    """
    plt.figure(figsize=(8, 10))


    # to graph multiple lines, you want to call plt.plot() as many times as needed before you call plt.show()
    plt.plot(x, y, linestyle=line_style, color=color, marker=marker)

    plt.hold on 
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plot_graph(x, y, title="Prime Numbers", xlabel="Index", ylabel="Prime", color="green", marker="o")


# more notes

""" 
np.arange([start, ]stop, [step, ], dtype=None)

```
