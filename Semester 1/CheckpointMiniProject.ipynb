```python
'''
1. Set the index to the name column.
2. Print the name and calories columns for cereals that have more than 110 calories.
3. Print the name and sugars columns for cereals that have less than or equal to 5 grams of sugar.
4. Print the name, calories, and sugars columns for cereals that have more than 110 calories AND less than 5 grams of sugar.
5. Find the cereal with the highest fiber content.
6. Print the cereals with a sodium content that is higher than the mean sodium content but have a rating below 50.
7. Find cereals that have a potassium content in the top 25% of all cereals and contain more than 5 grams of fiber.
8. Find cereals that have more than 100 calories and a shelf placement of 1 or 3.
9. Rank the cereals based on their rating, then print the top 5 cereals.

Bonus: Create a new column called "sugar_to_fiber_ratio" and calculate the ratio of sugars to fiber for each cereal. 
Then, print the cereals with a ratio greater than 2.

'''
```

```python
import pandas as pd

# Load the cereals dataset

cereals = pd.read_csv("cereal.csv")


#1. Set the index to the name column.

print(cereals.set_index("name"))

#2. Print the name and calories columns for cereals that have more than 110 calories.

print("2. Cereals with calories greater than 110")
print("----------------------------------------------------")
high_calorie_cereals = cereals[cereals["calories"] > 110]
print(high_calorie_cereals)
print()
print()

#3. Print the name and sugars columns for cereals that have less than or equal to 5 grams of sugar.

print("3. Cereals with sugars less than or equal to 5 grams")
print("----------------------------------------------------")
low_sugar_cereals = cereals[cereals["sugars"] <= 5]
print(low_sugar_cereals[["name", "sugars"]])
print()

#4. Print the name, calories, and sugars columns for cereals that have more than 110 calories AND less than 5 grams of sugar.

print("4. Cereals with more than 110 calories AND less than 5 grams of sugar")
print("----------------------------------------------------")
filtered_cereals = cereals[(cereals["calories"] > 110) & (cereals["sugars"] < 5)]
print(filtered_cereals[["name", "calories", "sugars"]])
print()

#5. Find the cereal with the highest fiber content.

print("5. Cereal with the highest fiber content")
print("----------------------------------------------------")
highest_fiber_cereal = cereals[cereals["fiber"] == cereals["fiber"].max()]
print(highest_fiber_cereal[["name", "fiber"]])
print()

#6. Print the cereals with a sodium content that is higher than the mean sodium content but have a rating below 50.

print("6. Cereals with sodium higher than mean and rating below 50")
print("----------------------------------------------------")
mean_sodium = cereals["sodium"].mean()
filtered_sodium_rating = cereals[(cereals["sodium"] > mean_sodium) & (cereals["rating"] < 50)]
print(filtered_sodium_rating[["name", "sodium", "rating"]])
print()

#7. Find cereals that have a potassium content in the top 25% of all cereals and contain more than 5 grams of fiber.

print("7. Cereals with potassium in the top 25% and fiber > 5 grams")
print("----------------------------------------------------")
threshold_potassium = cereals["potass"].quantile(0.75)
top_potassium_fiber = cereals[(cereals["potass"] > threshold_potassium) & (cereals["fiber"] > 5)]
print(top_potassium_fiber[["name", "potass", "fiber"]])
print()

#8. Find cereals that have more than 100 calories and a shelf placement of 1 or 3.

print("8. Cereals with more than 100 calories and shelf placement of 1 or 3")
print("----------------------------------------------------")
filtered_calories_shelf = cereals[(cereals["calories"] > 100) & (cereals["shelf"].isin([1, 3]))]
print(filtered_calories_shelf[["name", "calories", "shelf"]])
print()

#9. Rank the cereals based on their rating, then print the top 5 cereals.

print("9. Top 5 cereals based on rating")
print("----------------------------------------------------")
top_5_cereals = cereals.sort_values(by="rating", ascending=False).head(5)
print(top_5_cereals[["name", "rating"]])
print()

#Bonus: Create a new column called "sugar_to_fiber_ratio" and calculate the ratio of sugars to fiber for each cereal. 
#Then, print the cereals with a ratio greater than 2.
```
